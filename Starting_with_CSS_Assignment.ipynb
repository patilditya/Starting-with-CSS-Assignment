```python
# Question 1 : Create a simple page with some div tags and show different ways to add CSS as well as what happens when you target the same elements with inline, internal, and external CSS. Also, utilize comments in the project where required



--> HTML File



   <!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>CSS Demonstration</title>

    <link rel="stylesheet" href="CSS1.css">

    <!-- Internal CSS: CSS rules defined within the <style> tag in the <head> section -->
    <style>
        .internal-css {
            background-color: lightblue;
            color: white;
            padding: 10px;
            text-align: center;
        }
    </style>
</head>
<body>

    <!-- Inline CSS: Directly applied inside the HTML tag using the 'style' attribute -->
    <div style="background-color: lightgreen; color: white; padding: 10px; text-align: center;">
        <!-- This div uses inline CSS -->
        <p>Inline CSS Example</p>
    </div>

    <!-- External CSS: Targeting the div using a class from an external stylesheet (link below) -->
    <div class="external-css">
        <!-- This div uses external CSS -->
        <p>External CSS Example</p>
    </div>

    <!-- Internal CSS Example -->
    <div class="internal-css">
        <!-- This div uses internal CSS -->
        <p>Internal CSS Example</p>
    </div>

    <!-- External CSS Link -->
    <link rel="stylesheet" href="styles.css">

    <!-- Comments explaining what happens when we target the same elements with different CSS methods -->
    <p>
        <!-- The first div uses inline CSS, which will override internal or external styles,
             if there are conflicting styles. Inline styles have the highest specificity. -->
        Inline CSS takes priority over internal or external styles.
    </p>
    <p>
        <!-- The second div uses external CSS from the linked stylesheet (styles.css), which
             has a lower priority than inline styles but higher than internal styles if no conflicts arise. -->
        External CSS is useful for reusing styles across multiple pages.
    </p>
    <p>
        <!-- The third div uses internal CSS. This method is useful for a single page and
             provides better organization compared to inline CSS. Internal CSS is overridden by inline CSS. -->
        Internal CSS is used for styling on a specific page, but it has lower priority than inline CSS.
    </p>

</body>
</html>





--> CSS File


     /* External CSS targeting the same div as the internal CSS example */
.external-css {
    background-color: lightcoral;
    color: white;
    padding: 10px;
    text-align: center;
}

```

```python
# Question 2 : Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and explain how you've named the classes

-->
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>BEM Class Styling Example</title>

    <!-- Internal CSS -->
    <style>
        /* Block: .paragraph - styling for the base block */
        .paragraph {
            font-family: Arial, sans-serif;
            line-height: 1.6;
            margin-bottom: 20px;
        }

        /* Element: .paragraph__text - text inside the paragraph block */
        .paragraph__text {
            font-size: 16px;
            color: #333;
        }

        /* Modifier: .paragraph--highlight - applied to paragraphs that need highlighted style */
        .paragraph--highlight {
            background-color: yellow;
            font-weight: bold;
        }

        /* Modifier: .paragraph--important - applied to paragraphs that need extra emphasis */
        .paragraph--important {
            font-size: 18px;
            color: red;
        }

        /* Modifier: .paragraph--small - applied to small text paragraphs */
        .paragraph--small .paragraph__text {
            font-size: 14px;
            color: grey;
        }
    </style>
</head>
<body>

    <!-- Block: paragraph, with different modifiers -->
    <div class="paragraph paragraph--highlight">
        <p class="paragraph__text">This is a highlighted paragraph using the 'paragraph--highlight' modifier.</p>
    </div>

    <div class="paragraph paragraph--important">
        <p class="paragraph__text">This is an important paragraph using the 'paragraph--important' modifier.</p>
    </div>

    <div class="paragraph paragraph--small">
        <p class="paragraph__text">This is a small paragraph using the 'paragraph--small' modifier.</p>
    </div>

    <div class="paragraph">
        <p class="paragraph__text">This is a regular paragraph with no modifier.</p>
    </div>

</body>
</html>
```

```python
#Question 3 : Develop an HTML form with various input elements. Use CSS to style the form, including setting background colors for input fields. Create a custom color palette for the form elements, and demonstrate how to apply opacity to one of the form sections.

--> <!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Styled Form with Custom Colors</title>

    <!-- Internal CSS for styling the form -->
    <style>
        /* General body styling */
        body {
            font-family: Arial, sans-serif;
            background-color: #f4f4f4;
            padding: 20px;
        }

        /* Form container styling */
        .form-container {
            background-color: #fff;
            border-radius: 8px;
            box-shadow: 0 4px 8px rgba(0, 0, 0, 0.1);
            padding: 30px;
            width: 400px;
            margin: 0 auto;
        }

        /* Header styling */
        .form-container h2 {
            text-align: center;
            color: #2c3e50;
        }

        /* Label styling */
        label {
            display: block;
            margin-bottom: 10px;
            font-size: 16px;
            color: #2c3e50;
        }

        /* Input fields and textarea */
        input[type="text"], input[type="email"], input[type="password"], textarea {
            width: 100%;
            padding: 10px;
            margin: 10px 0 20px 0;
            border: 2px solid #ccc;
            border-radius: 5px;
            font-size: 16px;
        }

        /* Custom color palette for input fields */
        input[type="text"] {
            background-color: #e8f5e9; /* Light Green */
        }

        input[type="email"] {
            background-color: #e3f2fd; /* Light Blue */
        }

        input[type="password"] {
            background-color: #fff3e0; /* Light Orange */
        }

        textarea {
            background-color: #f1f8e9; /* Light Lime */
            resize: vertical;
        }

        /* Submit button styling */
        input[type="submit"] {
            background-color: #3498db; /* Blue */
            color: white;
            border: none;
            padding: 10px 20px;
            border-radius: 5px;
            font-size: 16px;
            cursor: pointer;
            width: 100%;
        }

        /* Hover effect for submit button */
        input[type="submit"]:hover {
            background-color: #2980b9;
        }

        /* Styling for form section with opacity */
        .form-section-opacity {
            opacity: 0.8; /* Applying opacity to this section */
            background-color: #ecf0f1; /* Light gray background */
            padding: 20px;
            border-radius: 5px;
        }

        /* Styling for error messages */
        .error-message {
            color: red;
            font-size: 14px;
            margin-bottom: 10px;
        }
    </style>
</head>
<body>

    <!-- Form container -->
    <div class="form-container">
        <h2>Contact Us</h2>

        <!-- Form section with opacity applied -->
        <div class="form-section-opacity">
            <form action="" method="post">
                <!-- Name input field -->
                <label for="name">Name</label>
                <input type="text" id="name" name="name" placeholder="Enter your name" required>

                <!-- Email input field -->
                <label for="email">Email</label>
                <input type="email" id="email" name="email" placeholder="Enter your email" required>

                <!-- Password input field -->
                <label for="password">Password</label>
                <input type="password" id="password" name="password" placeholder="Enter your password" required>

                <!-- Message textarea -->
                <label for="message">Message</label>
                <textarea id="message" name="message" placeholder="Enter your message" rows="4" required></textarea>

                <!-- Error message for demonstration -->
                <div class="error-message">Please fill out all fields!</div>

                <!-- Submit button -->
                <input type="submit" value="Submit">
            </form>
        </div>
    </div>

</body>
</html>
```
